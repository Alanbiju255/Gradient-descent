```python
import numpy as np
```

```python
print("ax^3 +bx^2+ cx +d =0 \n enter the values of a,b,c,d :\n")
a=int(input())
b=int(input())
c=int(input())
d=int(input())
```

```python

x=np.random.rand()
learningrate=0.01
epo=100
```

```python

for  epo in range (epo):
    value=a*x**3+b*x**2+c*x+d
    gradient=3*a*x**2+2*b*x+c
    if gradient<-1:
        gradient=0;
    x-=learningrate*gradient
```

```python
print(f"The equation is: {a}x^3 + {b}x^2 + {c}x + {d} = 0")
```

```python
print(f"The solution for x is approx. : {x}")
```
